```python
features = [
    'POS',
    'Billing',
    'Inventory',
    'Analytics',
    'Loyalty',
    'CRM',
    'QR',
    'Payments',
    'Multi Branch'
]

for feature in features:
    df[feature] = df['Key Features'].str.contains(
        feature,
        case=False,
        na=False
    ).astype(int)
```

```python
feature_counts = {}

for feature in features:
    feature_counts[feature] = df[feature].sum()

feature_df = pd.DataFrame(
    list(feature_counts.items()),
    columns=['Feature','Count']
)

fig = px.bar(
    feature_df,
    x='Feature',
    y='Count',
    text='Count',
    title='Feature Availability Across Competitors'
)

fig.show()
```

```python
opportunities = pd.DataFrame({
    'Opportunity':[
        'WhatsApp Ordering',
        'Bakery Workflow',
        'AI Analytics',
        'Loyalty Engine',
        'Inventory Prediction'
    ],
    'Demand':[9,10,8,7,8],
    'Competition':[3,2,5,6,4]
})

opportunities['Score'] = (
    opportunities['Demand'] -
    opportunities['Competition']
)

opportunities
```

```python
fig = px.bar(
    opportunities,
    x='Opportunity',
    y='Score',
    text='Score',
    title='ServeUp Opportunity Scorecard'
)

fig.show()
```
